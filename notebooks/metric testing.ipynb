```python
def compute_omega_b_m(pars):
    params   = pars[:,:6]
    h        = pars[:,2]/100
    omegabh2 = pars[:,0]
    omegach2 = pars[:,1]

    omeganh2 = (3.046/3)**(3/4)*0.06/94.1

    omegamh2 = omegabh2+omegach2+omeganh2

    omegab = omegabh2/h**2
    omegam = omegamh2/h**2

    return(omegab,omegam)

def compute_omega_b_c(pars):
    params   = pars[:,:5]
    H0       = params[:,2]
    omegab   = params[:,3]
    omegam   = params[:,4]
    omeganh2 = (3.046/3)**(3/4)*0.06/94.1

    h = H0/100

    omegabh2 = omegab*(h**2)
    omegach2 = (omegam-omegab)*(h**2) - omeganh2

    return(omegabh2,omegach2)

planck_chain = getdist.mcsamples.loadMCSamples('/home/grads/data/evan/cosmopower_emcee/notebooks/planck_emulated',no_cache=True)
joint_chain  = getdist.mcsamples.loadMCSamples('/home/grads/extra_data/evan/fiducial_chains/joint_lkl_0',no_cache=True)
lsst_chain   = getdist.mcsamples.loadMCSamples('/home/grads/data/evan/emulator/lsst_at_planck_test_mobo',no_cache=True)


s = planck_chain.samples
s[...,2] = 100*s[...,2] 
omb,omm = compute_omega_b_m(s)
s[...,0]=omb
s[...,1]=omm
planck_chain.setSamples(s)
planck_chain.setParamNames(['omegab','omegam','H0','tau','ns','logAs','Aplanck'])

# s = joint_chain.samples
# print(s[0,:5])
# omb,omc = compute_omega_b_c(s)
# s[...,3] = omb
# s[...,4] = omc
# print(s[0,:5])
# joint_chain.setSamples(s)
# print(planck_chain.samples[0])

print('Q_UDM optimal cutoff = %.3f' % gaussian_tension.Q_UDM_get_cutoff( planck_chain, lsst_chain, joint_chain)[0])
cutoff = gaussian_tension.Q_UDM_get_cutoff( planck_chain, lsst_chain, joint_chain)[0]
Q_UDM, Q_UDM_dofs = gaussian_tension.Q_UDM( planck_chain, joint_chain,param_names=['logAs', 'ns','H0','omegab','omegam'] )
Q_UDM_P = scipy.stats.chi2.cdf(Q_UDM, Q_UDM_dofs)
Q_UDM_nsigma = util.from_confidence_to_sigma(Q_UDM_P)
print(f'Q_UDM = {Q_UDM:.2f}, dofs = {Q_UDM_dofs:2}, P = {Q_UDM_P:.5f}, nsigma = {Q_UDM_nsigma:.3f}')
```
